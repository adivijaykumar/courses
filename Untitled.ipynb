```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import *
```

```python
a = np.matrix(np.random.normal(0,1,size=(10,10)))
t_arr = np.linspace(0,20,40)
for t in t_arr:
    #print norm(expm(t * a))
    plt.loglog(t,norm(expm(t * a)),'ro')
sa = np.amax(np.real(eigvals(a)))
plt.plo!t(t_arr, np.exp(t_arr * sa))
```

```python
mu = 5.
sig = 2.
N = 10000
bins = np.linspace(0,10,10)
x = np.random.normal(mu,sig,N)
plt.hist((x-mu)**2/sig**2,histtype='step',density=True,bins=bins)
y = np.random.chisquare(1,N)
plt.hist(y,histtype='step',density=True,bins=bins)
plt.show()
```

```python
mu = 50.
sig = 5.
N = 10000
bins = np.linspace(30,60,10)
x = np.random.normal(mu,sig,N)
y = np.random.chisquare(1,N)
plt.hist((x-mu)**2,histtype='step',bins=bins)
plt.hist(sig**2 * y,histtype='step',bins=bins)
plt.show()
```

```python
x = np.linspace(0,10,3000)
plt.loglog(x,chi2.pdf(x,1))

mu = 5
sig = 3
N = 1000000
y = np.random.normal(mu,sig,N)
plt.hist((y-mu)**2/sig**2,density=True,bins=x)
plt.show()
```
