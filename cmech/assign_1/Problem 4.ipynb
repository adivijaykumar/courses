
## Van der Pol Oscillator
Equation is $\ddot{x} -\mu (1-x^2)\dot{x}+x = 0$, $\mu > 0$

```python
import numpy as np, scipy.integrate, matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
mu_arr = [0.,1.,1.5,2.,-2.,-1.5,-1.,-.5,-.0001]
initial_condns = np.random.uniform(-10,10,(50,2))
```

```python
for mu in mu_arr:
    def f(t,y):
        x = y[0]
        xd = y[1]
        return np.array([xd,mu*(1-x**2)*xd-x])
    plt.figure(figsize=(4,3))
    for ini in initial_condns:
        sol = scipy.integrate.solve_ivp(f,[0,1000],[ini[0],ini[1]],
                                        t_eval=np.linspace(0,500,10000))
        plt.plot(sol.y[0],sol.y[1])
    plt.show()
```
