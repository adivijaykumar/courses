```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

## Problem 1

### Part (b) - The problem is of a particle moving in potential $V(x) =  \alpha x^2/2 + \beta x^4/4$

```python
def V(alpha,beta,x):
    x = np.asarray(x)
    return alpha*x**2 + beta*x**4

params = [[3,-1],[-3,1],[0.25,0.25]]
x = np.linspace(-2,2,100)

plt.figure(figsize=(6,6))
plt.plot(x,[0]*len(x),'--')

for parameter in params:
    alpha, beta = parameter
    plt.plot(x,V(alpha,beta,x),label='alpha = %s, beta = %s'%(alpha,beta))
plt.xlabel('$x$')
plt.ylabel('$V(x)$')
plt.legend()
plt.show()
```
