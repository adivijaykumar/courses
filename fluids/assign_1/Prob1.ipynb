```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import *
plt.style.use('presentation')
```

```python
kb = 1.38e-23
d = 1e-10
L = 1e-5
```

```python
h = np.loadtxt('data_PT.txt',unpack=True)
h[0] = 0.3048*h[0]
h[1] = 3386.39*h[1]
h[2] = 5./9*(h[2] - 32) + 273
```

```python
T_p = kb*h[2]/h[1]/1.414/3.14/d**2/L

plt.figure(figsize=(8,6))
plt.plot(h[0]/1000,T_p)
plt.xlabel('Height (in km)')
plt.ylabel('Knudsen Number')
```
