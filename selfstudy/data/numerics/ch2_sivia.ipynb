```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
h = 0.265
stdh = 0.2
def flat_prior(x):
    x = np.asarray(x)
    return( np.ones(len(x)))

def gaussian_prior(x):
    x = np.asarray(x)
    return(norm.pdf(x, h, stdh))

def gaussian_prior_shift(x):
    x = np.asarray(x)
    return(norm.pdf(x, h+0.5, stdh-0.1))
```

```python
plt.figure(figsize=(10,7))
chead, ctail = 0,0
bins = np.linspace(0,1,10000)
p = flat_prior(bins)
N = 5000
for i in range(N):
    test = np.random.uniform(0,1)
    if test < h:
        chead += 1
    else:
        ctail += 1
    
    if i in [0,7,15,31,63,127,255,511,1023]:
        res_f = bins ** chead * (1 - bins)** ctail * p
        res_f = (res_f/np.sum(res_f))
        plt.plot(bins, res_f,label='%s'%(i+1))
plt.plot([h,h],[0, 0.003],'--')
plt.legend()
plt.title('Flat Prior')
plt.show()

plt.figure(figsize=(10,7))
chead, ctail = 0,0
bins = np.linspace(0,1,10000)
p = gaussian_prior(bins)
N = 5000
for i in range(N):
    test = np.random.uniform(0,1)
    if test < h:
        chead += 1
    else:
        ctail += 1
    
    if i in [0,7,15,31,63,127,255,511,1023]:
        res_f = bins ** chead * (1 - bins)** ctail * p
        res_f = (res_f/np.sum(res_f))
        plt.plot(bins, res_f,label='%s'%(i+1))
plt.plot([h,h],[0, 0.003],'--')
plt.title('Gaussian Prior')
plt.legend()
plt.show()

plt.figure(figsize=(10,7))
chead, ctail = 0,0
bins = np.linspace(0,1,10000)
p = gaussian_prior_shift(bins)
N = 5000
for i in range(N):
    test = np.random.uniform(0,1)
    if test < h:
        chead += 1
    else:
        ctail += 1
    
    if i in [0,7,15,31,63,127,255,511,1023]:
        res_f = bins ** chead * (1 - bins)** ctail * p
        res_f = (res_f/np.sum(res_f))
        plt.plot(bins, res_f,label='%s'%(i+1))
plt.plot([h,h],[0, 0.003],'--')
plt.title('Shifted Gaussian Prior')
plt.legend()
plt.show()
```
