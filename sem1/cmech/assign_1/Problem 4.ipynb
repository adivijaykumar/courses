
## Van der Pol Oscillator
Equation is $\ddot{x} -\mu (1-x^2)\dot{x}+x = 0$, $\mu > 0$

The $\dot{x}$ term is like a damping term, with $x$ dependent damping coefficient $\delta(x) = -\mu (1-x^2)$. For $\mu > 0$ and $|{x}|>1$, this term is positive and hence there will be dissipation in the system, causing $|x(t)| + |\dot{x}(t)|$ to be bounded in time. For $|x|<1$, the value of $|x(t)| + |\dot{x}(t)|$ will increase till it approaches the limit cycle.

Some phase plots are plotted below.

```python
import numpy as np, scipy.integrate, matplotlib.pyplot as plt
plt.style.use('presentation')
```

```python
mu_arr = [0.,1.,1.5,2]
initial_condns = np.random.uniform(-10,10,(50,2))
```

```python
for mu in mu_arr:
    def f(t,y):
        x = y[0]
        xd = y[1]
        return np.array([xd,mu*(1-x**2)*xd-x])
    #plt.figure(figsize=(4,3))
    for ini in initial_condns:
        sol = scipy.integrate.solve_ivp(f,[0,100],[ini[0],ini[1]],
                                        rtol = 1e-7, atol=1e-7,
                                        t_eval=np.linspace(0,100,10000))
        plt.plot(sol.y[0],sol.y[1])
        plt.title(r'$\mu=%s$'%mu)
        plt.xlabel(r'$x$')
        plt.ylabel(r'$\dot{x}$')
    plt.show()
```

For $\mu < 0$ and $|{x}|>1$, this term is negative and hence there will be feedback in the system, causing $|x(t)| + |\dot{x}(t)|$ to be not bounded in time. For $|x|<1$, the value of $|x(t)| + |\dot{x}(t)|$ will decrease till it comes to rest.

```python
mu_arr = [-2.0,-1.5,-1.0,-0.5,-0.0001]
initial_condns = np.random.uniform(-0.5,0.5,(50,2))
```

```python
for mu in mu_arr:
    def f(t,y):
        x = y[0]
        xd = y[1]
        return np.array([xd,mu*(1-x**2)*xd-x])
    #plt.figure(figsize=(4,3))
    for ini in initial_condns:
        sol = scipy.integrate.solve_ivp(f,[0,100],[ini[0],ini[1]],
                                        rtol = 1e-7, atol=1e-7,
                                        t_eval=np.linspace(0,100,10000))
        plt.plot(sol.y[0],sol.y[1])
        plt.title(r'$\mu=%s$'%mu)
        plt.xlabel(r'$x$')
        plt.ylabel(r'$\dot{x}$')
    plt.show()
```
