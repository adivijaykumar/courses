```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(('presentation'))
```

```python
alpha, gamma, delta, m = 3.,1.,0.5,1.

def phi(om):
    return np.arctan(delta*om/(alpha-m*om**2))
    
def A(om):
    return gamma/np.sqrt(delta**2 * om**2 + (alpha - m *omega**2)**2)

omega = np.linspace(0.3,6,100)

plt.plot(omega, (A(omega)))
plt.xlabel(r'$\omega$')
plt.ylabel(r'$A$')
plt.figure()
plt.plot(omega,phi(omega))
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\phi$')
```
