```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('presentation')
```

```python
m= 100.
n= 100.
k = 2
d = 2.
plt.figure(figsize=(15,10))
X, Y = np.mgrid[-k:k+2:45j, -k:k:45j]
U = m*X/(X**2 + Y**2) + n*(X-d)/((X-d)**2 + Y**2)
V = m*Y/(X**2 + Y**2) + n*Y/((X-d)**2 + Y**2)
R = np.sqrt(U**2 + V**2)
plt.plot(0,0,'o',label='source')
plt.plot(d,0,'o',label='another source')
plt.xticks(())
plt.yticks(())

plt.legend()
plt.quiver(X,Y,U,V,R,alpha=0.8)
#plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=1.)
plt.show()
```
