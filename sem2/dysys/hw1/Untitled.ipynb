```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fn(t, x0):
    return(np.asarray((x0) / np.sqrt(1 + 2 * x0 ** 2 * t)))

def fn1(x0):
    return(-1/2/x0**2)
```

```python
plt.figure(figsize=(6,4.5))
plt.title('Solution')
t = np.linspace(0,1,100)
plt.plot(t, fn(t,20), label=r'$ x_0 > 0$')
plt.plot(t, fn(t,-20), label=r'$ x_0 < 0$')
plt.plot(t, fn(t,0), label=r'$ x_0 = 0$')
plt.xticks([])
plt.yticks([])
plt.legend()
plt.ylabel(r'$x(t)$')
plt.xlabel(r'Time $t$')
```

```python
plt.figure(figsize=(6,4.5))
plt.title('Region of Existence of Solution')
x0 = np.linspace(0.8,50,1000)
plt.plot(x0,fn1(x0),'r-',label=r'$t=\frac{1}{2x_0^2}$')
plt.plot(-x0,fn1(-x0),'r-')
plt.plot([0,0],[0.5,-.8],'r-')
plt.ylabel(r'$t$')
plt.xlabel(r'$x_0$')
plt.xticks([])
plt.yticks([])
plt.legend()
```

```python
def fn1(x0, t):
    t = np.asarray(t)
    return((x0/np.abs(x0)) * np.sqrt( 1./( 1 - (x0 ** 2 - 1)/x0 ** 2 * np.exp(2 * t))))
```

```python
plt.figure(figsize=(8,6))

t = np.linspace(-2, 0.1,100)
t1 = np.linspace(-2, 2,100)
plt.plot(t, fn1(3,t), label=r'$x_0 > 1$')
plt.plot(t, fn1(-3,t), label=r'$x_0 < 1$')
plt.plot(t1, fn1(0.3,t), label=r'$1 >x_0 > 0$')
plt.plot(t1, fn1(-0.3,t), label=r'$0 > x_0 > -1$')

plt.plot([-2,2],[0,0], label=r'$x_0=0$')
plt.plot([-2,2],[1,1], label=r'$x_0=1$')
plt.plot([-2,2],[-1,-1], label=r'$x_0=-1$')

plt.legend()
plt.title('Solutions')
plt.xticks([0])
plt.yticks([0,1,-1])
plt.grid(True)
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')
```

```python
plt.figure(figsize=(6,4.5))
def fnplot(x0):
    return(-0.5 * np.log(1 - 1/x0**2))

x0 = np.linspace(1.001, 3, 80)

plt.plot(x0, fnplot(x0), 'b-',label=r'$t = - 0.5 \log \left(1 - \frac{1}{x_0^2}\right) $')

x0 =  (-1) * x0
plt.plot(x0, fnplot(x0), 'b-')
plt.ylim(-3.2,3.2)
plt.plot([-1,-1], [-3.2,3.2],'r--')
plt.plot([1,1], [-3.2,3.2],'r--')
plt.yticks([0])
plt.xticks([0,1,-1])

plt.title('Region of Existence of Solution')
plt.xlabel(r'$x_0$')
plt.ylabel(r'$t$')
plt.grid(True)
plt.legend()
```
