```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,2))
N = 20
x = np.linspace(-1,1,N)
plt.quiver(x,np.zeros(N), - x ** 3 ,np.zeros(N))
plt.xticks([])
plt.yticks([])
plt.title('Vector Field')
plt.xlabel(r'$x$')

#plt.colorbar()
```

```python
plt.figure(figsize=(7,2))
N = 20
x = np.linspace(-100,100,N)
plt.quiver(x,np.zeros(N), x*(x**2 -1),np.zeros(N))
#plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.xlabel(r'$x$')
plt.title('Vector Field')
```

```python
plt.figure(figsize=(7,5))
N = 20
x = np.linspace(-2,2,20)
y = np.linspace(-2,2,20)
x1,x2 = np.meshgrid(x, y)
plt.quiver(x1,x2, x2, -x1 + x1 ** 3)
#plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Vector Field')
```
